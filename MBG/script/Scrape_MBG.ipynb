# Bagian 1: Daftar Satuan Pelayanan Pemenuhan Gizi (SPPG) yang telah beroperasional
url : <https://www.bgn.go.id/operasional-sppg>

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from google.colab import data_table

resp = requests.get('https://www.bgn.go.id/operasional-sppg')
soup = BeautifulSoup(resp.text, 'html.parser')

thead = soup.find('thead')
th = thead.find_all('th')

tbody = soup.find('tbody')
tbody_tr = tbody.find_all('tr')

data=[]
for tr in tbody_tr:
    td = tr.find_all('td')
    data.append({
        f"{th[0].text}" : f"{td[0].text}",
        f"{th[1].text}" : f"{td[1].text}",
        f"{th[2].text}" : f"{td[2].text}",
        f"{th[3].text}" : f"{td[3].text}",
        f"{th[4].text}" : f"{td[4].text}",
        f"{th[5].text}" : f"{td[5].text}",
    })

df = pd.DataFrame(data)
df.to_csv('Daftar SPPG.csv')
data_table.DataTable(df, include_index=False, num_rows_per_page=20)
```

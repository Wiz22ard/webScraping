# Daya Tampung Program Studi SNBT Tahun 2025
url : <https://snpmb.bppp.kemdikbud.go.id/utbk-snbt/daftar-ptn-snbt>

```python
import pandas as pd
from google.colab import data_table
from bs4 import BeautifulSoup
import requests
```

```python
base_url = 'https://sidatagrun-public-1076756628210.asia-southeast2.run.app/ptn_sb.php'

items = {
    'PTN AKADEMIK':'',
    'PTN VOKASI':'?ptn=-2',
    'PTN KIN':'?ptn=-3'
}

snbt = []
for label, url in items.items():
    response = requests.get(base_url+url)
    print(f"{label} : {response.status_code}")
    soup = BeautifulSoup(response.text, 'html.parser')
    tbody = soup.find('tbody')
    for row in tbody.find_all('tr'):
        data = row.find_all('td')
        kode_ptn = data[1].text
        url2 = base_url+f'?ptn={kode_ptn[1:]}'
        response2 = requests.get(url2)
        soup2 = BeautifulSoup(response2.text, 'html.parser')
        tbody2 = soup2.find('tbody')
        for row2 in tbody2.find_all('tr'):
            data2 = row2.find_all('td')
            kode_prodi = data2[1].text
            snbt.append({
                'KATEGORI':label,
                'KODE PTN':kode_ptn,
                'PTN':data[2].find_all('a')[0].text,
                'KABUPATEN':data[3].text,
                'PROVINSI':data[4].text,
                'KODE PRODI':data2[1].text,
                'PRODI':data2[2].text,
                'JENJANG':data2[3].text,
                'DAYA TAMPUNG 2025':data2[4].text,
                'PEMINAT 2024':data2[5].text,
                'JENIS PORTOFOLIO':data2[6].text
            })
        print(f"PTN:{data[2].find_all('a')[0].text}")

df_snbt = pd.DataFrame(snbt)
```

```python
df_snbt.to_csv('Daya_Tampung_Program_Studi_SNBT_Tahun_2025.csv', index=False, encoding='utf-8-sig')
data_table.DataTable(df_snbt, num_rows_per_page=10)
```
